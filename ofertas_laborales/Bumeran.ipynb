```python

palabras_clave = [  "analista de datos",
    "data analyst",
    "cientifico de datos",
    "data scientist",
    "business intelligence",
    "bi",
    "data engineer",
    "ingeniero de datos",
    "programador",
    "desarrollador",
    "software engineer",
    "backend",
    "frontend",
    "full stack",
    "ingeniero de sistemas",
    "sistemas",
    "machine learning",
    "analytics"]
resultados_totales = []

for q in palabras_clave:
    print(f"\n--- Iniciando extracción para: {q} ---")
    
    # Reiniciar la página para cada nueva palabra clave
    page = 0
    json_data = {'filtros': [], 'query': q}
    
    while True:
        params['page'] = str(page)
        
        response = rq.post(url, json=json_data, headers=headers, params=params, cookies=cookies)
            
        data = response.json()
        empleos = data['content']
                
        #terminar el bucle si no existen más empleos
        if len(empleos) == 0:
            print(f"  Fin de resultados para '{q}'. Total páginas: {page}")
            break
                
        #agregar key con la palabra clave utilizada
        for empleo in empleos:
            empleo['palabra_clave'] = q
            resultados_totales.append(empleo)
                
        page += 1 #siguiente página
        sleep(1.5)


# 3. Conversión a DataFrame final
df_final = pd.DataFrame(resultados_totales)
```
