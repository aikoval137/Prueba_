```python
import requests as rq
from selenium import webdriver
from time import sleep
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

### Obtener url bases

```python
url = "https://www.computrabajo.com/"
```

```python
URLS_TRABAJO = []
palabras_buscar = ["analista de datos",
    "data analyst",
    "cientifico de datos",
    "data scientist",
    "business intelligence",
    "bi",
    "analytics",
    "data engineer",
    "ingeniero de datos",
    "machine learning",
    "inteligencia artificial",
    "python",
    "sql",
    "power bi",
    "etl",
    "backend",
    "frontend",
    "full stack",
    "software engineer",
    "ingeniero de sistemas",
    "practicante sistemas",
    "practicante de datos"]

for palabra in palabras_buscar:
    sleep(1.5)
    driver = webdriver.Chrome()
    
    driver.get(url)
    
    driver.find_element(by= "id", value = "searchText").send_keys(palabra)
    ruta = '//*[@id="searchAddress"]'
    
    driver.find_element(by= "xpath", value =ruta).click()
    sleep(0.5)
    ruta2 = '//*[@id="js_autocomplete_searchAddress"]/li[3]'
    driver.find_element(by= "xpath", value =ruta2).click()
    
    ruta3 = '//*[@id="searchButton"]'
    driver.find_element(by= "xpath", value =ruta3).click()
    
    sleep(2)
    URLS_TRABAJO.append(driver.current_url)
    driver.close()
    
```

### Iniciar con 4 urls

```python
base = "https://pe.computrabajo.com"

urls_trabajos = set()
nombre_trabajo = []
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36",
    "Accept-Language": "es-ES,es;q=0.9,en;q=0.8"} 

for url in URLS_TRABAJO[:4]:
    print(f"SCRAPEANDO: {url} ")
    i=1
    while True: 
        sleep(2)
        url_pagina = f"{url}?p={i}"
       
        r = rq.get(url_pagina, headers = headers)
        if r.status_code == 403:
            print("bloqueo")
            break
        
        soup = bs(r.text)

        trabajos = soup.find_all('a', {'class': "js-o-link fc_base"})

        if len(trabajos)==0:
            break
        
        for trabajo in trabajos:
            href = base+str(trabajo.get("href"))
            if href not in urls_trabajos:
                urls_trabajos.add(href)
                nombre_trabajo.append(trabajo.text)
        print(f"pagina{i}")
        i+=1
    sleep(2.5)
```

### Continuar con las demĂˇs

### Obtener informaciĂłn de las urls filtradas

```python
trabajos = []

for index, url in enumerate(urls_filtradas):
    try:
        print(f"scrapeando {index}")
        r = rq.get(url, headers=headers)
        soup = bs(r.text)
        
        # tĂ­tulo y detalle del trabajo
        titulo_completo = soup.title.text
        detalle = soup.find('p', {'class': 'mbB'}).text

        if " - " in titulo_completo:
            partes = titulo_completo.split(" - ")
            titulo = partes[0]
            ubicacion = partes[-1]
        else:
            titulo = titulo_completo
            ubicacion = "No especificada"
            
        trabajos.append({
            'titulo': titulo,
            'ubicacion': ubicacion,
            'detalle': detalle,
            'portal': 'Computrabajo'
        })

        if (index + 1) % 50 == 0:
            pd.DataFrame(trabajos).to_csv("archivo_proceso.csv", index=False)
            print(f"Progreso: {index + 1} ofertas procesadas. Backup guardado")
        sleep(1.5)
    except Exception as e:
        print(e)
        
```

```python
df = pd.DataFrame(trabajos).to_csv("trabajos_computrabajo.csv", index=False)
```
